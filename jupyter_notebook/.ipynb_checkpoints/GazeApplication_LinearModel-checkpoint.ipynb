```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from PIL import Image
from sklearn import preprocessing
```

```python
#データの処理
#生データ
csv_data1 = pd.read_csv('./csvFile/さき_10-25-10-18_eyegaze_test.csv')
csv_data2 = pd.read_csv('./csvFile/さき_10-25-10-20_eyegaze_test.csv')
# csv_data_ = pd.concat([csv_data1, csv_data2])
csv_data_ = csv_data1
#必要なデータのみ取り出し
csv_data = csv_data_.drop(['frameId', 'timestamp', 'lookAtPosition_x', 'lookAtPosition_y'],axis=1)
csv_data = csv_data.drop(['face_0y', 'face_2z', 'right_1z', 'right_2y'],axis=1)
# #ターゲットと顔，右目，左目のデータ
# face_data = csv_data_.iloc[:, 0:18]
# data_ = csv_data_.iloc[:, 0:4]
# right_data_ = csv_data_.iloc[:, 19:34]
# right_data = pd.concat([data_, right_data_], axis=1)
# left_data_ = csv_data_.iloc[:, 35:50]
# left_data = pd.concat([data_, left_data_], axis=1)
#特徴量xとターゲットy
csv_data_x = csv_data.iloc[:,2:30]
csv_data_y = csv_data.iloc[:,0:2]
csv_data_x
```

```python
#データのプロット
plt.scatter(csv_data["target_x"], csv_data["face_0x"])
```

```python
# mscaler = preprocessing.MinMaxScaler()
# mscaler.fit(csv_data_x)
# xms = mscaler.transform(csv_data_x)
# mscaler.fit(csv_data_y)
# yms = mscaler.transform(csv_data_y)
sscalerx = preprocessing.StandardScaler()
sscalerx.fit(csv_data_x)
xss_sk = sscalerx.transform(csv_data_x) 
sscalery = preprocessing.StandardScaler()
sscalery.fit(csv_data_y)
yss_sk = sscalery.transform(csv_data_y)
```

```python
# トレーニング・テストデータ分割
X = csv_data_x
Y = csv_data_y
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
sscalerx = preprocessing.StandardScaler()
sscalerx.fit(X_train)
X_train = sscalerx.transform(X_train) 
X_test = sscalerx.transform(X_test) 
```

```python
#線形回帰（最小二乗法）
lr = LinearRegression() 
lr.fit(X_train, Y_train)
#予測　
lr_pred = lr.predict(X_test)
#評価
# 平均絶対誤差(MAE)
mae = mean_absolute_error(Y_test, lr_pred)
# 平方根平均二乗誤差（RMSE）
rmse = np.sqrt(mean_squared_error(Y_test, lr_pred))
# スコア
score = lr.score(X_test, Y_test)
print("MAE = %.2f, RMSE = %.2f, score = %.2f" % (mae, rmse, score)) 
lr_pred_ = sscalery.inverse_transform(lr.predict(X_test))
# print("Coef = ", lr.coef_) 
# print("Intercept =", lr.intercept_)

# #予測と正解の比較 
# print("Y_test:", Y_test[:5]) 
# print("Y_pred:", Y_pred[:5])

# #グラフ化　右肩上がりにデータが集まっていることが正しい予測を表す 
# plt.scatter(Y_test, Y_pred)
```

```python
#Lasso
rgs1 = Lasso(alpha=1) 

rgs1.fit(X_train, Y_train)#家の大きさ, 売却価格の関係を学習.
rgs1_pred = rgs1.predict(X_test)

#評価
# 平均絶対誤差(MAE)
mae = mean_absolute_error(Y_test, rgs1_pred)
# 平方根平均二乗誤差（RMSE）
rmse = np.sqrt(mean_squared_error(Y_test, rgs1_pred))
# スコア
score = rgs1.score(X_test, Y_test)
print("MAE = %.2f, RMSE = %.2f, score = %.2f" % (mae, rmse, score)) 
rgs1_pred_ = sscalery.inverse_transform(rgs1.predict(X_test))
```

```python
#Ridge
ridge= linear_model.Ridge(alpha=1.0)
ridge.fit(X_train, Y_train)
ridge_pred = ridge.predict(X_test)

#評価
# 平均絶対誤差(MAE)
mae = mean_absolute_error(Y_test, ridge_pred)
# 平方根平均二乗誤差（RMSE）
rmse = np.sqrt(mean_squared_error(Y_test, ridge_pred))
# スコア
score = ridge.score(X_test, Y_test)
print("MAE = %.2f, RMSE = %.2f, score = %.2f" % (mae, rmse, score)) 
```

```python
#ElasticNet
er= linear_model.ElasticNet(alpha=0.1, l1_ratio=0.5)
er.fit(X_train, Y_train)
er_pred = er.predict(X_test)

#評価
# 平均絶対誤差(MAE)
mae = mean_absolute_error(Y_test, er_pred)
# 平方根平均二乗誤差（RMSE）
rmse = np.sqrt(mean_squared_error(Y_test, er_pred))
# スコア
score = er.score(X_test, Y_test)
print("MAE = %.2f, RMSE = %.2f, score = %.2f" % (mae, rmse, score)) 
```
