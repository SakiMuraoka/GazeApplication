```python
#いろんな分類器を比較
names = ["Logistic Regression", "Nearest Neighbors", 
         "Linear SVM", "Polynomial SVM", "RBF SVM", "Sigmoid SVM", 
         "Decision Tree","Random Forest", "AdaBoost", "Naive Bayes", 
         "Linear Discriminant Analysis","Quadratic Discriminant Analysis"]

classifiers = [
    LogisticRegression(),
    KNeighborsClassifier(),
    SVC(kernel="linear"),
    SVC(kernel="poly"),
    SVC(kernel="rbf"),
    SVC(kernel="sigmoid"),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GaussianNB(),
    LDA(),
    QDA()]
resultx = []
resulty = []
for name, clf in zip(names, classifiers): # 指定した複数の分類機を順番に呼び出す
    clfx = clf
    clfy = clf
    clfx.fit(X_train, Y_train.iloc[:,[0]])
    clfy.fit(X_train, Y_train.iloc[:,[1]])
    score1x = clfx.score(X_train, Y_train.iloc[:,[0]]) # 正解率（train）の算出
    score2x = clfx.score(X_test, Y_test.iloc[:,[0]]) # 正解率（test）の算出
    score1y = clfx.score(X_train, Y_train.iloc[:,[1]]) # 正解率（train）の算出
    score2y = clfx.score(X_test, Y_test.iloc[:,[1]]) # 正解率（test）の算出
    resultx.append([score1x, score2x]) # 結果の格納
    resulty.append([score1y, score2y]) # 結果の格納

# test の正解率の大きい順に並べる
# df_result = pd.DataFrame(result, columns=['train', 'test'], index=names).sort('test', ascending=False)
# もし上のコードが動かない場合、以下のコーを試してみてください。
df_resultx = pd.DataFrame(resultx, columns=['trainx', 'testx'], index=names).sort_values('testx', ascending=False)
df_resulty = pd.DataFrame(resulty, columns=['trainy', 'testy'], index=names).sort_values('testy', ascending=False)
df_resultx # 結果の確認
# df_resulty # 結果の確認
# df_resultx.plot(kind='bar', alpha=0.5, grid=True)
# df_resulty.plot(kind='bar', alpha=0.5, grid=True)
```
